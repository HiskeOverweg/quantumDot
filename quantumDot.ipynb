# Quantum dot intro

```python
fig = plt.figure()
fig.set_size_inches(6,6)
ax = fig.add_subplot(111)
qd.fixed(ax)
plt.text(-1,3.2,'barrier',ha='center');
plt.text(1,3.2,'barrier',ha='center');
plt.plot(0,0,marker='o', color='k', ms = 8)
plt.text(0,0.2,'electron',ha='center',size=20);
```

This is a quantum dot - a small region in space in which an electron can be confined.

```python
fig = plt.figure()
fig.set_size_inches(6,6)
ax = fig.add_subplot(111)
qd.fixed(ax)
heights = qd.levels(ax,0)
plt.plot(0,0,'ko', ms = 8);
plt.text(-2,0.5,r'energy $\rightarrow$',rotation=90,va='center')
plt.text(1.1,1.9,r'$\rightarrow$ allowed level',size=20);
```

Electrons in a quantum dot occupy discrete energy levels, that is what is 'quantum' about them.

```python
fig = plt.figure()
fig.set_size_inches(6,6)
ax = fig.add_subplot(111)
qd.fixed(ax)
sd = qd.reservoirs(ax,0)
heights = qd.levels(ax,0)
[plt.plot(0,-i,'ko',ms=8) for i in range(0,3)];
```

The dot can be coupled to two large reservoirs, in which a continuum of states is occupied. The states in the dot are filled up to the same level.

```python
def plotter(bias):
    b = qd.ypos(bias)
    fig = plt.figure()
    fig.set_size_inches(6,6)
    ax0 = fig.add_subplot(111)
    qd.fixed(ax0)
    heights = qd.levels(ax0,0)
    [plt.plot(0,-i,'ko',ms=8) for i in range(0,3)]
    sd = qd.reservoirs(ax0,b)
    rectangle = mpatches.Rectangle((-1,-b*0.5),2,b,fc='springgreen',ec='none',alpha=0.2)
    ax0.add_patch(rectangle)
    plt.show()

    
plotter(5)
```

A voltage bias to the reservoirs leads to a so-called bias window.

```python
def plotter(bias,gate):
    g = qd.xpos(gate)
    b = qd.ypos(bias)
    fig = plt.figure()
    fig.set_size_inches(15,6)
    gs = gridspec.GridSpec(1, 2, width_ratios=[3, 5])
    gs.update( wspace= 0.3)
    ax0 = plt.subplot(gs[0])
    qd.fixed(ax0)
    sd = qd.reservoirs(ax0,b)
    heights = qd.levels(ax0,g)
    qd.carriers(ax0,heights,sd)
    ax1 = plt.subplot(gs[1])
    qd.measurement(ax1)
    plt.plot(gate,bias*1E-3,'o',ms=8,color='springgreen')
    plt.show()

    
plotter(5,-10.4)
```

Whenever a dot level is in the bias window, current can flow through the dot. This is what we measure.
